# Method

* Liner model is used to make prediction using old data by making a model like Y = wx+c . It is supervised learning, we have lable for given data. Liner model methods are simple to implement and we can avoid overfiting. Liner model has two type of method

 1. Least Square : In this method first we find weight using    
$$
\begin{align}
w &= \big(X^\top X\big)^{-1} X^\top T
\end{align}
$$

Then we will predict Y using formula 
$$
\begin{align}
Y &= w^\top x
\end{align}
$$

It is very simple and easy method to implement but the biggest issue of this method is memory cosumption. If data is bigger then it will be very hard to execute this method. For example here the data which I am using has 45,730 rows.So, there will be 45,730 row matrix and it will be very hard to compute inverse matrix. For small data this method is preferable.

 2. Least Mean Square: The second method of liner regression model. It is quite same, the only difference here is we are computing current weight using previous error, alpha rate and error. So we are updating our weight using error we get when we used previous weight. 
  * It does not have memory problem as least square. One of the main benefit of using LMS over LS.
  * Here we will get different weight for different alpha rate. We choose alpha rate by try and run method, generally start with  0.1. If you put alpha rate very small then there will be chance of overfitting.
  * To compute weight we use

$$
w^{(k+1)} = w^{(k)} - alpha \Big( {w^{(k)}}^\top x_k - t_k \Big) x_k.
$$




## Code Implemetation: 

* First we inheritent Abstract base class in our Basemodelclass. In that we define abstract method train and use. It is abstract method, we just pass in defination. So, we can overwrite in parent class.
* After that we make a class of Liner_model which inheritent BaseModel class. Here we initialized w to none in _init_ method and we will use this weight in our liner model. Then check_matrix is just for checking that it is 2-d matrix. Add-ones is for adding bias means here we are adding ones in new first column. Train and use is abstract method which we just pass.
* Least square is pretty simple, we are just adding bias and find w in train method. First we will take data and target as argument. In use, we are using w which we find in train method and return dot product of data and w. Which is our predicted output.
* Least Mean square we made three method use,train and train_step.The new method train_step is for updating weight from alpha rate,old weight and error. The main problem which I face is weight cannot be none first time. So, I used flag for assigning weight zero for first time and then I make flag true. That's why it will not run again. Now in train method we will call train_step method for every instance of data. Using that weight we will find predicted value in use method. 

### Super Classs Definition

```python
import numpy as np 
from abc import ABC, abstractmethod

# Super class for machine learning models 

class BaseModel(ABC):
    """ Super class for ITCS Machine Learning Class"""
    
    @abstractmethod
    def train(self, X, T):
        pass

    @abstractmethod
    def use(self, X):
        pass

    
class LinearModel(BaseModel):
    """
        Abstract class for a linear model 
        
        Attributes
        ==========
        w       ndarray
                weight vector/matrix
    """

    def __init__(self):
        """
            weight vector w is initialized as None
        """
        self.w = None

    def _check_matrix(self, mat, name):
        if len(mat.shape) != 2:
            raise ValueError(''.join(["Wrong matrix ", name]))
        
    # add a basis
    def add_ones(self, X):
        """
            add a column basis to X input matrix
        """
        self._check_matrix(X, 'X')
        return np.hstack((np.ones((X.shape[0], 1)), X))

    ####################################################
    #### abstract funcitons ############################
    @abstractmethod
    def train(self, X, T):
        """
            train linear model
            
            parameters
            -----------
            X     2d array
                  input data
            T     2d array
                  target labels
        """        
        pass
    
    @abstractmethod
    def use(self, X):
        """
            apply the learned model to input X
            
            parameters
            ----------
            X     2d array
                  input data
            
        """        
        pass 
```

### Least Squares

```python
# Linear Regression Class for least squares
class LinearRegress(LinearModel): 
    """ 
        LinearRegress class 
        
        attributes
        ===========
        w    nd.array  (column vector/matrix)
             weights
    """
    def __init__(self):
        LinearModel.__init__(self)
        
    # train lease-squares model
    def train(self, X, T):
        X1 = super().add_ones(X)
        self.w = np.linalg.inv(X1.T @ X1) @ X1.T @ T 
        
    
    # apply the learned model to data X
    def use(self, X):
        X1 = super().add_ones(X)
        return X1@(self.w)
        
```

### Least Mean Squares

```python
import collections # for checking iterable instance

# LMS class 
class LMS(LinearModel):
    """
        Lease Mean Squares. online learning algorithm
    
        attributes
        ==========
        w        nd.array
                 weight matrix
        alpha    float
                 learning rate
    """
    def __init__(self, alpha):
        LinearModel.__init__(self)
        self.alpha = alpha
        self.flag = False
    
    # batch training by using train_step function
    def train(self, X, T):
        for i in range(X.shape[0]):
            self.train_step(X[i],T[i])
         
            
    # train LMS model one step 
    # here the x is 1d vector
    def train_step(self, x, t):
        x1 = super().add_ones(x.reshape((1,x.shape[0])))
        if self.flag == False:
            self.w = np.zeros(x1.shape[1])
            self.flag = True
        y = self.w @  x1[0] 
        self.w -= self.alpha * (y - t) * x1[0]    
    
    # apply the current model to data X
    def use(self, X):
        X1 = super().add_ones(X)     
#         print((X1@self.w).reshape((X.shape[0],1)).shape)
        return (X1@self.w).reshape((X.shape[0],1))    
```
